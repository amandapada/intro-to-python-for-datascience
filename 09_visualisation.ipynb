## Visualisation

Human brains are terrible at processing arrays of raw numbers but highly optimized for pattern recognition. Visualisation is essentially data compression for the visual cortex.

In data science and AI, we visualize to do three things:

- Discover anomalies or errors in our datasets before feeding them to models.

- Understand the underlying probability distributions.

- Identify signals (features) that have strong predictive power.

Every visualization problem falls into one of these five classes:

1. Distribution (How is the data spread out?)
The goal is to see the shape, center, and spread of a single variable, and to spot outliers. This visualisation is usually done with 1 continuous numerical variable (e.g., Salary, Age, Distance). Examples of distribution charts are Histogram, Density Plot (KDE), Box Plot.

        Scenario: Analyzing the distribution of daily commute distances for dispatch riders in Lagos.  

2. Relationship (How do variables interact?)
To determine if one variable changes as another changes (correlation). This is the foundation of feature selection in machine learning. These types of plots need 2 Continuous Numerical variables. Primary charts include Scatter Plot, Bubble Chart, Heatmap (for multiple variables).

        Scenario: Determining if there is a relationship between a generator's running_hours and its maintenance_cost_naira.

3. Comparison (How do categories stack up against each other?)
We use these to evaluate the magnitude of differences between distinct groups. The required data types are 1 Categorical variable (the groups) AND 1 Numerical variable (the metric being measured). Bar Chart (Vertical or Horizontal) is an example of a comparison chart.

        Scenario: Comparing the average monthly data consumption (GB) across different telecom networks (MTN, Airtel, Glo, 9mobile).

4. Trend / Time-Series (How do things change over time?)
Used to track movements, seasonality, or cycles across a continuous timeline. Required Data Types are 1 Sequential/Time variable (Dates, Years, Timestamps) AND 1 Continuous Numerical variable. Line Chart, Area Chart are used for time-series analysis

        Scenario: Tracking the Naira-to-Dollar exchange rate over the last 365 days to identify volatility spikes.

5. Composition (What is the whole made of?)
These charts show how individual parts make up a total. 1 Categorical variable AND 1 Numerical variable (usually representing counts or percentages that sum to 100%) are required for this type of visualisation. Primary Charts are Stacked Bar Chart, Treemap. (Note: Avoid Pie Charts unless you only have 2 or 3 categories, as humans struggle to estimate angles accurately).

        Scenario: Visualizing the revenue breakdown of an e-commerce startup by product category (Electronics, Fashion, Groceries).

#### Matplotlib and Seaborn

#### Common Mistakes & Ethical Visualization
- Using a Line Chart for Categories: Connecting discrete categories (like Apple and Orange) with a line implies progression where none exists.

- Misleading Scales (Truncated Y-Axis): Starting the Y-axis at 50 instead of 0 to make a tiny 1% increase look like a massive spike. Explain how this is used to manipulate data in media and politics.

- The Pie Chart Trap: Human eyes are terrible at comparing angles. Use Bar charts instead of Pie charts in 95% of scenarios.

- Zero-Baseline Rule: Bar charts must ALWAYS start at 0 on the Y-axis. Truncating the axis visually exaggerates small differences, which is a common analytical shortcut and highly misleading.

- Color Blindness: Don't rely solely on Red/Green to indicate Bad/Good. Use varying textures, shapes, or colorblind-friendly palettes (like viridis).

- Data-Ink Ratio: A concept by Edward Tufte. Remove any ink (gridlines, borders, 3D effects) that does not directly present data. Keep it clean.
