### Feature Engineering

```python
# Create a new column
df['Age'] = 2024 - df['Year']
df[['Year', 'Age']].head(3)
```

```python
# Vectorized math operations
df['Price_per_HP'] = df['MSRP'] / df['Engine HP']
df[['MSRP', 'Engine HP', 'Price_per_HP']].head(3)
```

```python
# Binning: cut() for equal-width bins
df['HP_Category'] = pd.cut(df['Engine HP'], bins=[0, 150, 250, 350], labels=['Low', 'Medium', 'High'])
df[['Engine HP', 'HP_Category']].head(3)
```

```python
# Binning: qcut() for equal-frequency bins (quantiles)
df['MSRP_Quartile'] = pd.qcut(df['MSRP'], q=4, labels=['Q1', 'Q2', 'Q3', 'Q4'])
df[['MSRP', 'MSRP_Quartile']].head(3)
```

### Cleaning & Preprocessing

```python
# Drop rows with missing values
df.dropna()
```

```python
# Fill missing values with a specific value
df.fillna(0)
```

```python
# Convert column to different data type
df.astype({'Year': 'int64', 'MSRP': 'float64'})
```

```python
# Rename columns
df.rename(columns={'Engine HP': 'Horsepower', 'MSRP': 'Price'})
```

```python
# Remove duplicate rows
df.drop_duplicates()
```

```python
# Clean and standardize strings
df['Vehicle_Style'] = df['Vehicle_Style'].str.strip().str.lower()
```

### Grouping

```python
# Using agg() for multiple aggregations
df.groupby("Transmission Type")["MSRP"].agg(['mean', 'min', 'max', 'count'])
```

```python
# Multi-level aggregations
df.groupby(["Transmission Type", "Make"])["MSRP"].mean()
```

### Time Series

```python
# Convert to datetime
dates = pd.Series(['2020-01-15', '2021-03-20', '2022-06-10'])
pd.to_datetime(dates)
```

```python
# Create a datetime index and resample
ts_df = pd.DataFrame({'value': [10, 20, 30, 40, 50]}, 
                      index=pd.date_range('2024-01-01', periods=5, freq='D'))
ts_df.resample('2D').mean()
```

### Reshaping

```python
# Pivot: reshape data from long to wide format
pivot_df = df.pivot_table(values='MSRP', index='Make', columns='Transmission Type')
pivot_df
```

```python
# Melt: reshape data from wide to long format
melted = df[['Make', 'Model', 'Year', 'MSRP']].head(2)
pd.melt(melted, id_vars=['Make', 'Model'], value_vars=['Year', 'MSRP'])
```

### Merging & Joining

```python
# Merge (SQL-style joins)
df1 = df[['Make', 'Model', 'MSRP']].head(3)
df2 = df[['Make', 'Model', 'Year']].head(3)
pd.merge(df1, df2, on=['Make', 'Model'])
```

```python
# Concatenate (stacking) dataframes
df_part1 = df.iloc[0:2]
df_part2 = df.iloc[2:4]
pd.concat([df_part1, df_part2], ignore_index=True)
```
